## Lens Formula

$\frac{1}{a}+\frac{1}{b}=\frac{1}{f}$

## 렌즈의 공식을 그래프로 그리기

```python
# 볼록렌즈 (f=50mm)
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(12, 12))
a = list(range(0,50))+list(range(51,300))
b = [50*x/(x-50) for x in a]
plt.ylim(-1000, 1000)
plt.plot(a, b, 'k-')
plt.axvline(x = 100, color = 'b', linestyle='-.', label = "a=b=2f (m=1)")
plt.axhline(y = 50, color = 'r', linestyle='-.', label = "b = f")
text(120, 350, "Reduced Inverted Real Image (a > 2f)", rotation=0, verticalalignment='center')
text(55, 600, "Enlarged Inverted Real Image (f < a <= 2f)", rotation=0, verticalalignment='center')
text(5, -400, "Enlarged Erect Virtual Image (a < f)", rotation=0, verticalalignment='center')
plt.grid(True)
plt.legend(loc=1)
```

## 볼록 렌즈에 의한 상의 작도 (Ray Diagram of Convex Lens)

```python
import matplotlib.patches as patches

plt.figure(figsize=(12, 12))
X=np.array([-150, 150])
Y=np.array([-80,80])
plt.plot(X,Y,color='None')
plt.ylim(-50, 80)

## 렌즈 그리기
shp=patches.Ellipse((0,0), 10, 50, color='g',angle=0)
plt.gca().add_patch(shp)
# 초점 그리기
shp=patches.Circle((-50,0), 1, color='b')
plt.gca().add_patch(shp)
text(-52, -4, "f", rotation=0, verticalalignment='center')
shp=patches.Circle((50,0), 1, color='b')
plt.gca().add_patch(shp)
text(48, -4, "f", rotation=0, verticalalignment='center')

## 1. 초점 거리 안(a=30)에 물체가 있을 때 (돋보기)
# 물체 그리기
shp=patches.Rectangle((-30-2,0), 4, 20, color='k')
plt.gca().add_patch(shp)
text(-40, 24, "Object-1", rotation=0, verticalalignment='center')
# 두 개의 광선으로 상의 위치 그리기
plt.plot([0, 60], [20, -4],color="b") # 평행한 광선
plt.plot([-30, 60], [20, -40],color="b") # 렌즈의 중심를 통과하는 광선
plt.plot([-100, 0], [60, 20], "b--") # 허상으로부터 나오는 광선 그리기
plt.plot([-100, -30], [200/3, 20], "b--") # 허상으로부터 나오는 광선 그리기
# 상 그리기
shp=patches.Rectangle((-75-2,0), 4, 50, color='gray') #, fill=None)
plt.gca().add_patch(shp)
text(-90, 54, "Image-1(Virtual)", rotation=0, verticalalignment='center')

## 2. 초점 거리의 두 배 거리(a=2f=100)에 물체가 있을 때 (m=1)
# 물체 그리기
shp=patches.Rectangle((-100-2,0), 4, 20, color='magenta')
plt.gca().add_patch(shp)
text(-120, 24, "Object-2", rotation=0, verticalalignment='center')
# 상의 위치 작도
plt.plot([-124, 0], [20, 20],color="magenta") # 평행한 광선-1
plt.plot([0, 125], [20, -30],color="magenta") # 평행한 광선-2
plt.plot([-125, 125], [25, -25],color="magenta") # 렌즈의 중심를 통과하는 광선
# 상 그리기
shp=patches.Rectangle((100-2,0), 4, -20, color='pink') #, fill=None)
plt.gca().add_patch(shp)
text(80, -24, "Image-2(Real)", rotation=0, verticalalignment='center')

plt.grid(True)
```
